3. Input kolom data yang dibutuhkan pada x dan y ke variabel sesuai dengan pemodelan dan analisis yang diinginkan. Pada kasus ini, data x berupa kolom Jumlah pertandingan yang dimainkan, dan umur. Sedangkan data y berupa Jumlah Assist Lionel Messi.

```
feature_cols = ['Matches Played', 'Age']
X = df[feature_cols]
y = df.Assists # Variabel Target
```

4. Dataset dipecah menjadi training set(x) dan test set(y). Alasannya, jika kumpulan data dipecah menjadi training set dan test set, tidak terdapat cukup data dalam kumpulan data training set untuk model mempelajari pemetaan yang efektif dari input ke output. Data juga tidak akan cukup dalam test set untuk mengevaluasi performa model secara efektif.

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

5. Pada tahap ini, dilakukan **Pembuatan Decision Tree/Pohon Keputusan menggunakan Library scikit-learn**. Tahap ini menjelaskan mengenai klasifikasi suatu objek serta memprediksi respon dari test dataset.

```
# Membuat decision tree berdasarkan objek
clf = DecisionTreeClassifier()

# Train Decision Tree berdasarkan pengklasifikasian (berisi pecahan dataset pada langkah sebelumnya)
clf = clf.fit(X_train,y_train)

# Melakukan prediksi respon dari test dataset
y_pred = clf.predict(X_test)
```

<h3>EVALUASI MODEL</h3>

---
Dilakukan tahap evaluasi tekait hasil pada Tahap Pemodelan Data sebelumnya






6. Melakukan print hasil dari Nilai Akurasi Model, jadi seberapa sering pengklasifikasi benar? Pada kasus saya tadi dilakukan Analisis Deskriptif terkait Jumlah assist Lionel Messi dalam mempengaruhi Umur dan Jumlah pertandingan yang ia mainkan. Akurasi model menunjukan **nilai tingkat klasifikasi 14,28% yang akan dianggap sebagai akurasi yang baik**.

```
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
